```python
js = {"dynamic typing", "interpreted", "web native", "prototype oriented"}
python = {"dynamic typing", "interpreted", "scientific computing", "object oriented"}
java = {"compiled", "object oriented", "static typing"}

langages = [js, python, java]
```

```python
# Your code here
```

### Solution

```python
print(f"common features of JavaScript and Python :                            {js & python}")
print(f"exclusive characteristics of Java compared to Python :                {java - python}")
print(f"full set of characteristics across all languages :                    {set.union(*langages)}")
print(f"full set of characteristics across all languages :                    {set.union(js, python, java)}")
print(f"exclusive characteristics of Python compared to all other languages : {python.difference(js, java)}")
```

- How to access the second month of winter?

### Solution

```python
winter = ["December", "January", "February"]
spring = ["March", "April", "May"]
summer = ["June", "July", "August"]
autumn = ["September", "October", "November"]

seasons = {
    "winter": winter,
    "spring": spring,
    "summer": summer,
    "autumn": autumn
}
print(seasons["winter"][1])
print("sumMer" in seasons)
```
